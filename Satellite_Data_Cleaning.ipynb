```python
import pandas as pd
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 500)
df = pd.read_excel('UCS_Satellite_Database_9-1-2017.xlsx')
df.head()
```

```python
df1 = df.dropna(axis = 1, how = 'all') #drop columns that are only NA values and save data frame as df1
df1.info()
```

```python
df1.rename(columns = {'Name of Satellite, Alternate Names':'Name','Country/Org of UN Registry':'RegistryCountry', 'Detailed Purpose':'DetailedPurpose','Perigee (km)':'Perigee(km)','Apogee (km)':'Apogee(km)','Inclination (degrees)':'Inclination(degrees)','Period (minutes)':'Period(minutes)','Launch Mass (kg.)':'LaunchMass(kg)','Dry Mass (kg.)':'DryMass(kg)','Power (watts)':'Power(W)','Date of Launch':'DateofLaunch','Country of Operator/Owner':'CountryofOwner','Class of Orbit':'ClassofOrbit','Type of Orbit':'TypeofOrbit','Expected Lifetime':'ExpectedLifetime', 'Country of Contractor':'CountryofContractor','Launch Site':'LaunchSite', 'Launch Vehicle':'LaunchVehicle','COSPAR Number':'COSPAR#','NORAD Number':'NORAD#'}, inplace = True)
#Rename column names for easy accesibility
```

```python
df1.head()
```

```python
df1['Name'].value_counts(dropna = False) #Find duplicated data in Satellite names column
```

```python
df1.drop_duplicates('Name', inplace=True) # Remove duplicate data only leaving the first copy
```

```python
df1.info()
```

```python
df1.drop(['DetailedPurpose','Comments','Unnamed: 27','Source Used for Orbital Data','Source','Source.1','Source.2','Source.3','Source.4','Source.5','Unnamed: 58','Unnamed: 59','Unnamed: 60','Unnamed: 64'], axis=1, inplace=True)
#Drop columns that will not be used in data analysis
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1['Users'].value_counts(dropna=False)
```

```python
df1.replace(to_replace={'Commercial/Government':'Government/Commercial', 'Commercial/Military': 'Military/Commercial', 'Government/Military':'Military/Government','Civil/Government':'Government/Civil','Military ':'Military', 'Commerical':'Commercial'}, inplace=True)
```

```python
df1['Users'].value_counts(dropna=False)
```

```python
for x in range(df1['Users'].count()):
    if df1['Users'].get(x) == 'Commercial/Government/Military':
        df1['Users'].drop(x, inplace=True)
```

```python
df1['Users'].value_counts(dropna=False)
```

```python
df1['CountryofOwner'].value_counts(dropna=False)
```

```python
names = ['Taiwan/USA','USA/Argentina','USA/Japan','France/Italy','USA/Canada','USA/Russia','Germany/USA','India/France','USA/France','France/Italy/Belgium/Spain/Greece','France/Belgium/Sweden','Russia/USA','Greece/United Kingdom','Singapore/Taiwan','ESA/USA','India/Canada','USA/United Kingdom/Italy','Russia/Multinational','China/Brazil','United Kingdom/ESA','Singapore/Japan','Ukraine/China','USA/Canada/Japan','France/USA','ESA/USA/Russia','Turkmenistan/Monaco','USA/Japan/Brazil','Turkey/France','France/Israel','Morocco/Germany']
```

```python
for x in range(df1['CountryofOwner'].count()):
    if df1['CountryofOwner'].get(x) in names:
        df1['CountryofOwner'][x] = 'Multinational'
```

```python
df1['CountryofOwner'].value_counts(dropna=False)
```

```python
df1['CountryofOwner'].replace({'Russia ':'Russia'}, inplace=True)
```

```python
df1['CountryofOwner'].value_counts(dropna=False)
```

```python
df1['Purpose'].value_counts(dropna=False)
```

```python
purpose_list = ['Communications/Technology Development','Earth Observation/Technology Development','Communications/Maritime Tracking','Earth Observation/Communications','Earth Science/Space Science','Space Science/Technology Development','Communications/Navigation','Earth Observation/Space Science','Earth Observation/Communications/Space Science','Technology Development/Educational','Earth Science/Earth Observation']
dict1 = dict.fromkeys(purpose_list,'Multipurpose')
```

```python
df1['Purpose'].replace(to_replace=dict1, inplace=True)
```

```python
df1['Purpose'].replace({'Earth Observation ':'Earth Observation', 'Navigation/Global Positioning':'Navigation', 'Navigation/Regional Positioning':'Navigation'}, inplace = True)
```

```python
df1['Purpose'].value_counts(dropna=False)
```
