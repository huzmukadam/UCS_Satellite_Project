```python
country_list = np.array(df['CountryofOwner'])
users_list = np.array(df['Users'])
purpose_list = np.array(df['Purpose'])
classoforbit_list = np.array(df['ClassofOrbit'])
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
le.fit(purpose_list)
encoded_purpose = le.transform(purpose_list)
len(encoded_purpose)
```

```python
dummy_countries = pd.get_dummies(country_list)
dummy_users = pd.get_dummies(users_list)
dummy_classoforbit = pd.get_dummies(classoforbit_list)
```

```python
df_dummies = dummy_countries.join(dummy_users).join(dummy_classoforbit).join(pd.DataFrame(encoded_purpose))
```

```python
df_dummies.info()
```

```python
df['Purpose']
```

```python
df.info()
```

```python
df2 = df.merge(df_dummies, left_index=True, right_index=True)
df2.head()
```

```python
df2.drop(columns=['Purpose','Name', 'RegistryCountry', 'CountryofOwner','Operator/Owner','Users','ClassofOrbit','TypeofOrbit','DryMass(kg)','Power(W)','DateofLaunch','ExpectedLifetime','Contractor', 'CountryofContractor','LaunchSite', 'LaunchVehicle','COSPAR#','NORAD#'], inplace=True)
```

```python
df2.drop(columns = ['LaunchMass(kg)'], inplace=True)
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.dropna(inplace=True)
```

```python
df2.info()
```

```python
df2['Period(minutes)'].replace({'8 days':11520}, inplace=True)
```

```python
df2.rename(columns={0:'encoded_purpose'}, inplace=True)
```

```python
df2.info()
```

```python
X2 = df2.drop(columns='encoded_purpose')
y2 = np.array(df2['encoded_purpose'])
```

## Logistic Regresion

```python
# Split the data into a training and test set.
Xlr, Xtestlr, ylr, ytestlr = train_test_split(X2.iloc[:, 0:6], y2, random_state=5)

clf = LogisticRegression()
# Fit the model on the trainng data.
clf.fit(Xlr, ylr)
# Print the accuracy from the testing data.
print(accuracy_score(clf.predict(Xtestlr), ytestlr))
```

## KNN Classification

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size =0.33)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

knn.score(X_test, y_test)
```

## Confusion Matrix and Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
